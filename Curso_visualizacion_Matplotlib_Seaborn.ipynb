##Clase 5 - Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=4)
axes[0,0].plot(x,np.cos(x),'b')
axes[0,1].plot(x,np.sin(x),'r')
axes[0,2].plot(x,np.tan(x),'g')
axes[0,3].plot(x,np.cos(x)**2,'black')
axes[1,0].plot(x,np.sin(x)**2,'gray')
fig.tight_layout()
```

```python
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(nrows=2, ncols=4)
ax1.plot(x,np.cos(x),'b')
ax2.plot(x,np.sin(x),'r')
ax3.plot(x,np.tan(x),'g')
ax4.plot(x,np.cos(x)**2,'black')
ax5.plot(x,np.sin(x)**2,'gray')
fig.tight_layout()
```

##Clase 6 - Leyendas, etiquetas, títulos, tamaño

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
axes[0].plot(x,y,'b', label='$sin(x)$')
axes[0].set_title('Relacion x - y')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].legend(loc=0)
axes[1].plot(y,x,'r', label='$sin(y)$')
axes[1].set_title('Relacion y - x')
axes[1].set_xlabel('y')
axes[1].set_ylabel('x')
axes[1].legend(loc=0)
fig.tight_layout()
```

```python
plt.plot(x,y,label='$sin(x)$')
plt.title('Título')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.show()
```

##Clase 8 - Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['COLOMBIA', 'JAPAN', 'MEXICO', 'GERMANY', 'FRANCE']
population = [1000, 800, 1200, 900, 1300]
```

```python
plt.figure(figsize=(8, 4))
plt.bar(country, population, width=0.4, color=['green', 'blue'])
plt.xticks(np.arange(5),('Colombia', 'Japon', 'Mexico', 'Alemania', 'Francia'), rotation=45)
plt.title('Poblacion de paises')
plt.xlabel('Paises')
plt.ylabel('Poblacion')
plt.show()
```

```python
plt.barh(country, population, height=0.4, color=['green', 'blue'])
plt.show()
```

##Clase 15 - Relation

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size')
plt.legend(loc='center', bbox_to_anchor=(1.08, 0.5))
plt.show()
```

```python
plt.figure(figsize=(10, 6))
markers = {"Lunch": "D", "Dinner": "s"}
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers)
plt.legend(loc='center', bbox_to_anchor=(1.08, 0.5))
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=tips, x='total_bill', y='tip', hue='day', style='time')
plt.legend(loc='center', bbox_to_anchor=(1.08, 0.5))
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', kind='scatter', col='time')
plt.show()
```
